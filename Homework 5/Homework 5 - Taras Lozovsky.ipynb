**2.** Користувач вводить 10 цілих чисел. Виведіть сумму цих числе.

```
L = input("Введіть десять цілих чисел, розділяючи їх пробілом: ").split()

while len(L)<10:
    L += input(f"Введіть ще {10-len(L)} чисел: ").split()

S = 0
for i in range(10):
    S += int(L[i])

print(f"Сумма перших 10 введених чисел = {S}")
```

**3.** Заповніть список випадковими числами (довільна кількість). Виведіть лише парні числа та їх індекси

```
import random as rnd

L = rnd.choices(range(1,1000),k=10)

for i, v in enumerate(L):
    if v % 2 ==0:
        print(i, v)
```

**6.** Користувач вводить 10 числел. Знайдіть середне значення, медіану, середне квадратичне відхилення.

```
import math as mt

L = input("Введіть десять цілих чисел, розділяючи їх пробілом: ").split()

while len(L)<10:
    L += input(f"Введіть ще {10-len(L)} чисел: ").split()

N = list(map(int,L[:10]))
N.sort()

# Mean
S = 0
for i in N:
    S += i

Mean = S/10

# Median
Median = (N[4]+N[5])/2

# Standart deviation
S = 0
for i in N:
    S += (i - Mean)**2

Std = mt.sqrt(S/9)

print(f"Середне значення = {Mean}")
print(f"Медіана = {Median}")
print(f"Стандартне выдхилення = {Std}")
```

**7.** Користувач вводить 10 числел. Підрахувати кількість простих чисел із введених.

```
import math as mt

def isprime(n):
    """
    Return True if n is prime number
    """

    if n < 2:
        return False
    elif n == 2 or n == 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    
    for i in range(1,int(mt.sqrt(n)/6)+1):
        p = 6*i - 1
        q = 6*i + 1
        if n % p == 0 or n % q == 0:
            return False
    else:
        return True
    

L = input("Введіть десять цілих чисел, розділяючи їх пробілом: ").split()

while len(L)<10:
    L += input(f"Введіть ще {10-len(L)} чисел: ").split()

k = 0
for i in map(int,L[:10]):
    if isprime(i):
        k += 1

print(f"Введено {k} простих чисел.")
```
