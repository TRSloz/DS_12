### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```
import random
import statistics as st

def create_digits_dictinary(rnl):
    """
    Returns dictionary with digits in list of numbers 'RNL'
    """
    out = dict(zip(range(0,10), [0 for i in range(0,10)]))
    
    for i in rnl:
        if i < 10:
            out[i] += 1
        elif i < 100:
            out[i // 10] += 1
            out[i %  10] += 1
        elif i == 100:
            out[1] += 1
            out[0] += 2

    return out

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

MinRndList = min(list_with_random_numbers)
MaxRndList = max(list_with_random_numbers)
MeanRndLst = st.mean(list_with_random_numbers)
MedRndList = st.median(list_with_random_numbers)
StdRndList = st.stdev(list_with_random_numbers)

print(f"Мінімальне значення у послідовності:  {MinRndList}")
print(f"Максимальне значення у послідовності: {MaxRndList}")
print(f"Середне значення у послідовності:     {MeanRndLst}")
print(f"Медіана послідовності:                {MedRndList}")
print(f"Середнє квадратичне відхилення:       {StdRndList}")

TNumbDict = create_digits_dictinary(list_with_random_numbers)
print("Частота зустрічаємості цифр у послідовності:")
for i in TNumbDict:
    print(f"    Цифра - {i}: {TNumbDict[i]} разів")
```
