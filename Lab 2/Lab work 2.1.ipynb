#### Завдання 2

Дано трицифрове число. Виведіть окремо кожну цифру числа та суми цифр число

Реалізуйте завдання двома способами (використовуючи стрічки та оператори `%`, `//`)

```
N = input()

S = 0
for d in N:
    print(d)
    S += int(d)

print(f"Сумма цифр = {S}")
```

```
N = int(input())

S = 0
while N > 0:
    d = N % 10
    S += d
    print(d)
    N //= 10

print(f"Сумма цифр = {S}")
```

#### Завдання 5

Напишіть програму для обчислення коренів квадратного рівняння.

Нагадую, що `n`-го степеня має `n` коренів. (Корені можуть бути комплексні)

```
import math as mt
import cmath as cmt

a = int(input()) # x^2
b = int(input()) # x^1
c = int(input()) # x^0

D = cmt.sqrt( mt.pow(b,2) - 4*a*c )
X1 = (-b - D) / (2 * a)
X2 = (-b + D) / (2 * a)

print(f"Перший корінь рівняння: {X1}")
print(f"Другий корінь рівняння: {X2}")
```
