У файлі мітиться інформація про товарні позиції деякого інтернет-магазину, а саме:
* Ариткул - унікальний номер товарної позиції
* Найменування - найменування товару
* Наявність - чи є товарна позиція в даний момент
* Ціна - вартість товарної позиції

```
db = {}

for row in open('raw.txt'):
    tmp = row.split('|')
    db[tmp[0]] = [tmp[1], tmp[2], float(tmp[3].rstrip(' грн\n'))]
```

### Завдання 1

Вам потрібно отримати таку інформацію:

* кількість товарних позицій у файлі
* кількість товарних позицій, що в наявності
* ціна та назва найдорожчої товарної позиції

```
print(f"кількість товарних позицій у файлі - {len(db)}")
```

```
# by for-loop
qty = 0
for k, v in db.items():
    if v[1] == 'В наличии':
        qty += 1
print(f"a) кількість товарних позицій, що в наявності - {qty}")

# by list comprehension
tmp = [k for k,v in db.items() if v[1] == 'В наличии']
print(f"b) кількість товарних позицій, що в наявності - {len(tmp)}")

# by functional programing
tmp = filter(lambda k: db[k][1] == 'В наличии', db)
print(f"c) кількість товарних позицій, що в наявності - {len(list(tmp))}")
```

```
# by for-loop
max_price = 0
max_prod = None

for k,v in db.items():
    if max_price < v[2]:
        max_price = v[2]
        max_prod = k

print(f"ціна та назва найдорожчої товарної позиції: {db[max_prod][0]} - {max_price} грн")

# by lambda function
max_prod = max(db, key=lambda k: db[k][2])
print(f"ціна та назва найдорожчої товарної позиції: {db[max_prod][0]} - {db[max_prod][2]} грн")
```
