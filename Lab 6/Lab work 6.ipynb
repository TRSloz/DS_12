```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style('whitegrid')
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
fig = plt.figure(figsize=(10,8), dpi=100)

# plot 'a' series
ax1 = fig.add_subplot(221)
ax1.plot(df.a, label='df.a')
ax1.grid(False)
ax1.legend()

# plot 'b' series
ax2 = fig.add_subplot(222)
ax2.plot(df.b, label='df.b')
ax2.grid(False)
ax2.legend()

# plot 'c' series
ax3 = fig.add_subplot(223)
ax3.plot(df.a, label='df.c')
ax3.grid(False)
ax3.legend()

# plot y=x line
ax4 = fig.add_subplot(224)
ax4.plot(range(31))
ax4.grid(False)
```

Створити візуалізацію, аналогічно рисунку

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
sns.set_style('darkgrid')

fig4 = plt.figure(figsize=(10,5), dpi=100)

# plot 'a' series
ax9 = fig4.add_subplot()
ax9.plot(df.a, label='df.a')

# plot 'b' series
ax9.plot(df.b, label='df.b')
ax9.grid(True)
ax9.legend()
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
fig5 = plt.figure(figsize=(10,5), dpi=100)

# plot 'a' series
ax10 = fig5.add_subplot()
ax10.plot(df.a, '-.r', label='df.a')

# plot 'b' series
ax10.plot(df.b, color='orange', linewidth =10, label='df.b')

# plot 'c' series
ax10.plot(df.c, 'y-o', linewidth =1, label='df.c')

ax10.grid(True)
ax10.legend()
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig6, axs11 = plt.subplots(3, 1, figsize=(10, 10), dpi=100)

# plot 'a' series
axs11[0].plot(df.a, label='Line A')
axs11[0].legend(loc='lower right')

# plot 'b' series
axs11[1].plot(df.b, label='Line B')
axs11[1].legend(loc='center left')

# plot 'c' series
axs11[2].plot(df.b, label='Line C')
axs11[2].legend()
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
fig7 = plt.figure(figsize=(10,5), dpi=100)

# plot 'a' series
ax12 = fig7.add_subplot()
ax12.bar(range(len(df.a)), df.a, 0.5, color='tomato')
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
import matplotlib.ticker as ticker

fig8 = plt.figure(figsize=(13,5), dpi=100)

# plot 'a' series
ax13 = fig8.add_subplot()
ax13.plot(df.a, '-^', label='')
ax13.set_ylabel('This is the Y Axis', fontsize=8)
ax13.set_xlabel('This is the X Axis', fontsize=14)
ax13.set_title('This is the Title', fontsize=8)
ax13.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax13.legend()
```
